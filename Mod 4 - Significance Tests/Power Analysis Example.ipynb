## [Power of Power Analysis]('https://www.analyticsvidhya.com/blog/2020/12/statistics-for-beginners-power-of-power-analysis/')

### [`tt_ind_solve_power` documentation]('https://www.statsmodels.org/stable/generated/statsmodels.stats.power.tt_ind_solve_power.html')


```python
from statsmodels.stats.power import TTestIndPower
import numpy as np

effec_size = 0.9
alpha_sig = 0.05
power = 0.9

# load and pass the input parameters to calc the sample size
power_obj = TTestIndPower()
sample_size = power_obj.solve_power(effect_size=effec_size, alpha = alpha_sig, power = power)
print('SAMPLE SIZE:', np.ceil(sample_size))
```

### Plot Power Curves to check how varying the effect size and sample size changes the power of the experiment

[`plot_power` documentation]('https://www.statsmodels.org/stable/generated/statsmodels.stats.power.TTestPower.plot_power.html')


```python
import matplotlib.pyplot as plt
power_obj.plot_power(nobs = np.arange(10,100), effect_size=[0.2, 0.5, 0.8], alpha=0.05)
plt.show()
```
