## 2.8 Probability Primer

Probability = numerical measure of the likelihood or chance that an event will occur

P(A) = probability of A occurring

### Probability Scale
- Scale from 0 to 1
- near 0 = unlikley to occur
- 0.5 = 50/50 chance
- near 1 = very likely
- 0 = **impossible event**
- 1 = **certain event**

### Assignming Probabilities


#### Classical Approach
- finite # of outcomes, each of which equally likely
- with 'n' possibilities, probability of 1/n for each outcome
- example rolling dice

#### Relative Frequency Approach
- data available to estimate the proportion of time the outcome will occur if we repeat experiment large number of times
- often used to determine probs for a particular population 
- P(A) = nA / n
    - nA = number of outcomes for A over total number of trials 'n'
    

#### 2 basic requirements for either approach
- probability of an outcome is between 0 and 1
- sum of probabilities of all outcomes = 1

### Addition Rule
P(A or B) = P(A) + P(B) - P(A and B)
if two events **mutually exclusive** - can't occur at same time
P(A or B) = P(A) + P(B)

#### Example:
pick a card from deck - what is prob of card being a jack(A) or a diamond(B)

P(A) = 4/52 and P(B) = 13/52

NOT mutually exclusive as card can be both a jack and a diamond

P(A and B) = 1 / 52

Soultion: P(A or B) = 4/52 + 13/52 - 1/52 = 16/52 = 4/13

### Complementary Events
IF: P(A) + P(B) = 1  && A intersect B = 0 => A and B complementary events a

Event A and its complement A' are complementary and: P(A') = 1 - P(A)

#### Example
P(A) = 0.4 // P(B) = 0.8 // P(A union B) = 0.9
Find:
- P(A intersect B):
    - P(A ∩ B) = P(A) + P(B) - P(A ∪ B)
    - 0.4 + 0.8 - 0.9 = 0.3
- P(A'):
    - P(A') = 1 - P(A) = 1 - 0.4 = 0.6
- P(A' union B): 
    - P(A’ ∪ B) = P(A’) + P(B) - P(A’ ∩ B)
    - P(A’ ∩ B) = P(B) - P(A ∩ B) = 0.8 - 0.3 = 0.5
    - P(A’ ∪ B) = 0.6 + 0.8 - 0.5 = 0.9
    
    
### Mutually Exclusive Events
Two events A and B are said to be mutually exclusive if they
cannot occur together:
- A ∩ B = 0
- The addition rule becomes:
    -P(A ∪ B) = P(A) + P(B)
- Event A and its complement A’ are mutually exclusive (i.e. complementary events are mutually exclusive):
    -A ∩ A’ = 0

## 2.12 Conditional Probability

P(A|B) = conditional probability of A given that B has occurred

Event B = **conditioning event**

P(A|B) = P(A ∩ B) / P(B) provided P(B) > 0

In terms of frequency:

**\# of outcomes in both A and B / \# of outcomes in B**

**Example:**

In a high school, 50% of the female students have black hair, 30% have blue eyes and 15% have both black hair and blue eyes. A female student is selected at random.

    a) Given that the student has black hair, find the probability that she has blue eyes.
        P(blue eyes | black hair) = P(black hair and blue eyes) / P(black hair)
        = 0.15 / 0.5 = 0.3
    b) Given that the student has blue eyes, find the probability that she has black hair.
        P(black hair | blue eyes) = P(blue eyes and black hair) / P(blue eyes)
        = 0.15 / 0.3 = 0.5
    c) Find the probability that the student has either black hair or blue eyes or both.
        P(A U B) = 0.35 + 0.15 + 0.15 = 0.65

### Multiplicative rule for probabilities
by-product of the definition of conditional probability

#### P(A∩B) = P(A┃B)\*P(B)
when A and B swapped
#### P(A∩B) = P(B┃A)\*P(A)

#### so P(A┃B)\*P(B) = P(B┃A)\*P(A)

Two way table to help work out different probabilities - check leture at 9:20 as this slide not in slides pdf

### Multiple stages in succession
- multiplicative rule most useful when experiment consists of several stages in succession
- extend rule to experiments involving more than two stages

**P(A1∩A2∩A3) = P(A3┃A1∩A2)P(A1∩A2) = P(A3┃A1∩A2)P(A2┃A1)P(A1)**

*Here A1 occurs first then A2 then A3*

### Independent Events

Two events A and B are independent if:
**P(A∩B) = P(A) x P(B)**

This means:

**P(A┃B) = P(A) and P(B┃A) = P(B)**

In other words, knowing that event B has occurred does not change the probability of A and vice versa
